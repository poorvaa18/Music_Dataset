```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #decisionTreeClassifier is class implements decision tree algo

from sklearn.model_selection import train_test_split #splitting the data into two sets- training and testing
from sklearn.metrics import accuracy_score
```

```python
Music_df=pd.read_csv('../input/beginner-ml-music-dataset/musicDataset.csv')
Music_df
```

```python

#we need to split the data into two sets. INPUT and OUTPUT to predict if we want to what 21 year old will like
X_df=Music_df.drop(columns=['genre'])
X_df
Y_df=Music_df['genre']
Y_df
X_df_train,X_df_test,Y_df_train,Y_df_test=train_test_split(X_df,Y_df,test_size=0.23)
```

```python
#creating a object - model with instance decisionTreeClassifier
model=DecisionTreeClassifier() 
model.fit(X_df_train,Y_df_train) #fit takes two dataset- training dataset
prediction=model.predict(X_df_test) #predict method takes in two dimensional array with new input dataset
prediction #we'll compare the x_test with y_test which is the actuall output
score=accuracy_score(Y_df_test,prediction)
score
```
